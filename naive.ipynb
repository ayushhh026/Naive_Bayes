```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# They follow continous numerical distributioon so we will use 
```

```python
# Gaussian Naive Bayes 
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
df['time']=df['time'].map({'Dinner':0,'Lunch':1})
```

```python
X=df.drop(columns=['time','size'])
y=df['time']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
clt=ColumnTransformer(
    transformers=(('ohe',OneHotEncoder(drop='first'),['sex','smoker']),('ord',OrdinalEncoder(),['day'])),
    remainder='passthrough'
)
```

```python
X_trans=clt.fit_transform(X_train)
X_test_trans=clt.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gag=GaussianNB()
gag.fit(X_trans,y_train)
```

```python
y_pred=gag.predict(X_test_trans)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
